```python
#import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn import metrics
import seaborn as sns
```

import adver.dataset

#import libraries

```python
#import data
data=pd.read_csv('Advertising.csv')
data.head
```

```python
#generate discriptive statistics
data.describe()
```

```python
data.info()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['TV'],y=data['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Radio'],y=data['Sales'])
plt.show()
plt.figure(figsize=(8,5))
sns.scatterplot(x=data['Newspaper'],y=data['Sales'])
plt.show()
```

```python
#splitdata
train,test=train_test_split(data,test_size=0.30)
```

```python
train_x=train[['TV','Radio','Newspaper']]
train_y=train.Sales

test_x=train[['TV','Radio','Newspaper']]
test_y=train.Sales
```

```python
train_x.head
```

```python
train_y.head
```

```python
#train model
model=linear_model.LinearRegression()
model.fit(train_x,train_y)
predct=model.predict(test_x)
metrics.r2_score(predct,test_y)
```

Regression is a mathematical analysis of dataset which help to prediction general Eq Y=A+BX

```python
#Regression Graph
plt.figure(figsize=(10,5))
sns.regplot(x=test_y,y=predct,scatter_kws={'color':'red'})
```

Note:- Regression shows the dataset in line. This trend line is called "BEST FIT OF LINE". In this dataset are some outlier which affect the analysis of model so, best analysis to remove the outlier from dataset firstly.
